```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tracks=pd.read_csv("tracks.csv")
tracks.head(100
           )
```

```python
#checking null values
pd.isnull(tracks).sum()
```

```python
tracks.info()
```

```python
sorteDf=tracks.sort_values('popularity',ascending=True).head(10)
```

```python
tracks.describe().transpose()
```

```python
mostPopular=tracks.query('popularity>90',inplace=False).sort_values('popularity',ascending=False)
mostPopular[:10]
```

```python
tracks.set_index("release_date", inplace=True)
tracks.index = pd.to_datetime(tracks.index)
tracks.head()
```

```python
tracks["duration"] = tracks["duration_ms"].apply(lambda x: round(x/1000))
tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
tracks
```

```python
tracks.duration.head()
```

```python
corr=tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidth=1,linecolor="Black")
heatmap.set_title("Correlation heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sampleDf=tracks.sample(int(0.004*len(tracks)))
```

```python
len(sampleDf)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sampleDf,y="loudness",x="energy",color="c").set(title="Loudness vs energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sampleDf,y="popularity",x="acousticness",color="b").set(title="Loudness vs energy")
```

```python
tracks['dates']=tracks.index.get_level_values('release_date')
tracks.dates=pd.to_datetime(tracks.dates)
years=tracks.dates.dt.year
```

```python
totalDr=tracks.duration
figDims=(18,7)
fig, ax= plt.subplots(figsize=figDims)
fig= sns.barplot(x=years,y=totalDr,ax=ax,errwidth=False).set(title="Years VS duration")
plt.xticks(rotation=90)
```

```python
totalDr=tracks.duration
sns.set_style(style="whitegrid")
figDims=(10,5)
fig, ax= plt.subplots(figsize=figDims)
fig= sns.lineplot(x=years,y=totalDr,ax=ax).set(title="Years VS duration")
plt.xticks(rotation=60)
```
